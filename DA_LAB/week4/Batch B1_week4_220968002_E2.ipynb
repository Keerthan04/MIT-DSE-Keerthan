# 18) Visualize the Toss decision across seasons

```python
toss_decision_counts = df.groupby(['season', 'toss_decision']).size().reset_index(name='count')
toss_decision_counts
```

```python

plt.figure(figsize=(12, 6))
sns.barplot(x='season', y='count', hue='toss_decision', data=toss_decision_counts)
plt.title('Toss Decision Across Seasons in IPL')
plt.xlabel('Season')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='Toss Decision', loc='upper right')
plt.show()
```

# 21) Visualize the Highest MOM award winners

# 23)Compare Total Played Matches vs Winning Matches vs Win Rate

```python
Total_played=pd.concat([df["team1"],df["team2"]])
a=Total_played.value_counts()
aa=a.tolist()
aa
dd=pd.DataFrame(aa, columns=["Total Played Matches"])
b=df.winner.value_counts()
bb=b.tolist()
ddd=pd.DataFrame(bb, columns=["Total won Matches"])
c=a.index.tolist()
d=pd.DataFrame(c, columns=["Name of team"])
e = pd.merge(d, dd, left_index=True, right_index=True)
e = pd.merge(e, ddd, left_index=True, right_index=True)
e["win rate"]=(e["Total won Matches"]/e["Total Played Matches"]*100)
e
```
