# EXERCISE 1: Data Analysis using CEREALS dataset

```python
import pandas as pd
df=pd.read_excel("Cereals1.xls")
df
```

# DATA Preprocessing first

```python
print("5-number summary of all the numeric attributes is:")
df.describe()
```

```python
for c in df.columns:
    if(c=="protein" or c=="fat" or c=="sodium" or c=="fiber" or c=="carbo" or c=="sugars" or c=="potass" or c=="vitamins"):
        mean=df[c].mean()
        df[c].replace(to_replace=-1, value=mean, inplace=True)
df
```

```python
df.describe()
```

```python
for c in df.columns:
    if(c=="protein" or c=="fat" or c=="sodium" or c=="fiber" or c=="carbo" or c=="sugars" or c=="potass" or c=="vitamins"):
        q3=df[c].quantile(0.75)
        q1=df[c].quantile(0.25)
        iqr=q3-q1
        uth=q3+(1.5*iqr)
        lth=q1-(1.5*iqr)
        median1=df[c].median()
        df[c]=df[c].apply(lambda x:median1 if x>uth or x<lth else x)
df
```

```python
df.describe()
```

# 5. Which manufacturers produce cereal with highest calories?

```python
df1=df.groupby(by="mfr")["calories"].sum()
for i in range(len(df1.index)):
    maximum=df1.max()
    if(df1[i]==maximum):
        print("the manufacturer with highest calories is ",df1.index[i])
```
