# 2)For each of the numeric attributes (proteins up to vitamins), identify and replace all missing data (indicated with -1) with the arithmetic mean of the attribute.

```python
for c in df.columns:
    if(c=="protein" or c=="fat" or c=="sodium" or c=="fiber" or c=="carbo" or c=="sugars" or c=="potass" or c=="vitamins"):
        mean=df[c].mean()
        df[c].replace(to_replace=-1, value=mean, inplace=True)
df
 
```

# For each of the numeric attributes (proteins up to vitamins), identify and replace all noisy data with the median of attribute. 

```python
for c in df.columns:
    if(c=="protein" or c=="fat" or c=="sodium" or c=="fiber" or c=="carbo" or c=="sugars" or c=="potass" or c=="vitamins"):
        q3=df[c].quantile(0.75)
        q1=df[c].quantile(0.25)
        iqr=q3-q1
        uth=q3+(1.5*iqr)
        lth=q1-(1.5*iqr)
        median1=df[c].median()
        df[c]=df[c].apply(lambda x:median1 if x>uth or x<lth else x)
df
```
