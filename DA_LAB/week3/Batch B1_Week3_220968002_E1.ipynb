# EXERCISE 1: Data Analysis using mtcars

```python
import pandas as pd
df=pd.read_csv("mtcars.csv")
df
```

# 1. Find the car with the best mpg

```python
best=df[df.mpg==df.mpg.max()].model
print("The car with the best mpg is ",best)
```

# 2. Find the car with the worst mpg

```python
worst=df[df.mpg==df.mpg.min()].model
print("The car with the worst mpg is ",worst)
```

# 3. Find the car with the best horsepower

```python
best=df[df.hp==df.hp.max()].model
print("The car with the best horsepower is ",best)
```

# 4. Find 5 number summary of displacement

```python
df.disp.describe()
```

# 5. Find median horse power

```python
med=df.hp.median()
print("median horse power is ",med)
```

# 6. What is average mpg for manual vs. automatic cars

```python
df1=df.groupby(by="am").mpg.mean()
df1
```

# 7.Draw a histogram of miles per gallon

# 8. Boxplot of mpg for each cylinder type

```python
sns.boxplot(x="cyl",y="mpg",data=df)
```
