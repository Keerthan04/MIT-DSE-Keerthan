#  2.Prepare the data for finding association rules. Each transaction will contain a list of item in the transaction

```python
my_dict={}
def dataextract():
    for i in range(0, 9509):
        a = []
        for j in range(0, 169):
            if not pd.isna(df.iloc[i, j]):
                a.append(df.iloc[i, j])
        my_dict[i] = a

dataextract()
#has a direct formula like list(csv.reader something check)
```

```python
my_dict
```

```python
my_list = list(my_dict.values())
my_list
```

# 3.Use Python library mlxtend and convert the transactions into a format that can be used in the Apriori method for finding frequent itemsets. 

```python
!pip install mlxtend
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
te = TransactionEncoder()
te_ary = te.fit(my_list).transform(my_list)
df1 = pd.DataFrame(te_ary, columns=te.columns_)
df1
```

# 4. Find top selling items with minimum support of 2%.

```python
percent2support=apriori(df1, min_support=0.02, use_colnames=True)
percent2support
```

```python
percent2support.sort_values(by="support",ascending=False)
```

# 5.Find all frequent itemsets with minimum support of 5%.

```python
frequent_itemsets = apriori(df1, min_support=0.05, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

# 6.Find all frequent itemsets of length 2 with minimum support of 2%.

```python
frequent_itemsets1 = apriori(df1, min_support=0.02, use_colnames=True)
frequent_itemsets1['length'] = frequent_itemsets1['itemsets'].apply(lambda x: len(x))
frequent_itemsets1[frequent_itemsets1['length'] == 2]
```

# 7.Find the top 10 association rules with minimum support of 2%, sorted by confidence in descending order.

```python
ar1=association_rules(frequent_itemsets1, metric="confidence",min_threshold=0.0)
ar1
```

```python
ar1.sort_values(by="confidence",ascending=False).head(10)
```

# 10.Find association rules with minimum support of 2% and lift of more than1.0

```python
ar2=association_rules(frequent_itemsets1, metric="confidence",min_threshold=0.0)
ar2
```

```python
ar2=ar2[ar2["lift"]>=1.0]
ar2
```
