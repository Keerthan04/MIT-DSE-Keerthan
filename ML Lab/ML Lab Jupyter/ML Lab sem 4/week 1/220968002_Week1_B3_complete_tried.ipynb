```python
corr
```

```python
sns.heatmap(annot=True, data=corr)
```

```python
for column in df.columns:
    q1=df[column].quantile(0.25)
    q3=df[column].quantile(0.75)
    iqr=q3-q1
    ul=q3+1.5*iqr
    ll=q1-1.5*iqr
    median=df[column].median()
    df[column]=df[column].where((df[column]>=ll) & (df[column]<=ul),other=median)
```

```python
df.head()
```

```python
plt.figure(figsize=(15,8))
sns.boxplot(data=df)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
data=pd.DataFrame(scaler.fit_transform(df))
```

```python
data
```

```python
cor=df.corr()
```

```python
cor
```

```python
sorted_mat = cor.unstack().sort_values() 
sorted_mat
```
