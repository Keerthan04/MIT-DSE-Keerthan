## Using Manual Method

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('excel.xlsx')
df
```

```python
x=np.array(df[['Weeks']])
y=np.array(df[['Sales In thoushands']])
```

```python
x
```

```python
y
```

```python
plt.title("Scatter diagram")
plt.scatter(x,y)
for i, j in zip(x, y):
 plt.annotate('(%s, %s)' % (i, j), xy=(i, j), textcoords='offset points', xytext=(0,10), ha='center')
plt.show()
```

```python
mean_x=np.mean(x)
mean_y=np.mean(y)
mean_x
```

```python
mean_y
```

```python
mean_x2=(np.sum(x*x))/len(x)
mean_x2
```

```python
mean_xy=np.sum(x*y)/len(x)
mean_xy
```

```python
b1=((mean_xy)-((mean_x)*(mean_y)))/((mean_x2)-((mean_x)*(mean_x)))
b1
```

```python
b0=mean_y-(b1*mean_x)
b0
```

```python
plt.title("scatter diagram with line after manual solving")
y_pred=b0+(x*b1)
plt.scatter(x,y)
plt.plot(x,y_pred,color="orange")
for i, j in zip(x, y):
 plt.annotate('(%s, %s)' % (i, j), xy=(i, j), textcoords='offset points', xytext=(0,10), ha='center')
plt.show()
```
