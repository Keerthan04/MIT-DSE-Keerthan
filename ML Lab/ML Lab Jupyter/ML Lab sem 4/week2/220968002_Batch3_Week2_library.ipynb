# Using Library Method

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_excel('excel.xlsx')
```

```python
df
```

```python
lr=LinearRegression()
```

```python
X=np.array(df['Weeks']).reshape(-1,1)#should be 2d array to send or use X=df[['Weeks']]
Y=np.array(df['Sales In thoushands']).reshape(-1,1)
```

```python
X
```

```python
Y
```

```python
plt.title("Scatter diagram")
plt.scatter(X,Y)
for i, j in zip(X, Y):
 plt.annotate('(%s, %s)' % (i, j), xy=(i, j), textcoords='offset points', xytext=(0,10), ha='center')
plt.show()
```

```python
lr.fit(X,Y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.title("scatter diagram with line")
y_pred=lr.intercept_+(X*lr.coef_)
plt.scatter(X,Y)
plt.plot(X,y_pred,color="orange")
for i, j in zip(X, Y):
 plt.annotate('(%s, %s)' % (i, j), xy=(i, j), textcoords='offset points', xytext=(0,10), ha='center')
plt.show()
```
